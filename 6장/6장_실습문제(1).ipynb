```python
address = data['공장대표주소']
address.str.contains('의창구').value_counts()
```

```python
import numpy as np
```

```python
gu = ['성산구', '의창구', '마산합포구', '마산회원구', '진해구']
num_factory = np.zeros(5)
for x in range(5):
    tmp = address.str.contains(gu[x]).value_counts()
    num_factory[x] = tmp[1]
    
np.sum(num_factory)
```

```python
Number_Factory_Gu = pd.Series(num_factory, index=gu)
Number_Factory_Gu
```
